```python
X=pokemon[['Total','HP','Attack','Defense','Sp_Atk','Sp_Def','Speed']]
y=pokemon.isLegendary
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
```

```python
model = RandomForestClassifier(max_leaf_nodes=10)
model.fit(X_train, y_train)
```

```python
normal=[[318,45,49,49,65,65,45]]
legend=[[580,90,85,100,95,125,85]]
print("Normal Pokemon",model.predict(normal))
print("Legendary Pokemon",model.predict(legend))
```
